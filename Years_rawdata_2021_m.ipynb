```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Step 1: Load Data
file_path = '/content/drive/MyDrive/bob_publication_data/years/2021/raw/2021_m/0.csv'  # Replace with the actual path to your CSV file
df = pd.read_csv(file_path)
```

```python

# Step 2: Remove Unnecessary Columns
columns_to_remove = ["X", "uni_time", "numeric.time"]
df_cleaned = df.drop(columns=columns_to_remove, errors='ignore')
```

```python
# Step 3: Summarize the Data
print("Summary of the data:")
print(df_cleaned.describe(include='all'))
```

```python

# Step 4: Visualize the Data

# Convert 'time' column to datetime for better plotting
df_cleaned['time'] = pd.to_datetime(df_cleaned['time'])
```

```python
# Plotting temperature inside the hive
plt.figure(figsize=(10, 6))
plt.plot(df_cleaned['time'], df_cleaned['t_i_1'], label='t_i_1')
plt.plot(df_cleaned['time'], df_cleaned['t_i_2'], label='t_i_2')
plt.plot(df_cleaned['time'], df_cleaned['t_i_3'], label='t_i_3')
plt.plot(df_cleaned['time'], df_cleaned['t_i_4'], label='t_i_4')
plt.plot(df_cleaned['time'], df_cleaned['t_i_5'], label='t_i_5')
plt.xlabel('Time')
plt.ylabel('Temperature (Â°C)')
plt.title('Temperature Inside the Hive Over Time')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Plotting outside temperature
plt.figure(figsize=(10, 6))
plt.plot(df_cleaned['time'], df_cleaned['t_o'], label='t_o', color='orange')
plt.xlabel('Time')
plt.ylabel('Outside Temperature (Â°C)')
plt.title('Outside Temperature Over Time')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Plotting hive weight
plt.figure(figsize=(10, 6))
plt.plot(df_cleaned['time'], df_cleaned['weight_kg'], label='Weight (kg)', color='green')
plt.xlabel('Time')
plt.ylabel('Weight (kg)')
plt.title('Hive Weight Over Time')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Plotting humidity
plt.figure(figsize=(10, 6))
plt.plot(df_cleaned['time'], df_cleaned['h'], label='Relative Humidity (%)', color='blue')
plt.xlabel('Time')
plt.ylabel('Relative Humidity (%)')
plt.title('Relative Humidity Over Time')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Plotting pressure
plt.figure(figsize=(10, 6))
plt.plot(df_cleaned['time'], df_cleaned['p'], label='Pressure (hPa)', color='purple')
plt.xlabel('Time')
plt.ylabel('Pressure (hPa)')
plt.title('Pressure Over Time')
plt.legend()
plt.grid(True)
plt.show()
```
